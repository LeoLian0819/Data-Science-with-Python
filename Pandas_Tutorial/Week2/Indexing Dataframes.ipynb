```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv')
df.head()
```

```python
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head()
```

```python
df.columns
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns = {col: 'Gold' + col[4:]}, inplace = True)
    if col[:2] == '02':
        df.rename(columns = {col: 'Silver' + col[4:]}, inplace = True)
    if col[:2] == '03':
        df.rename(columns = {col: 'Bronze' + col[4:]}, inplace = True)
    if col[:1] == '№':
        df.rename(columns = {col: '#' + col[2:]}, inplace = True)
df.head()
```

```python
# Boolean Mask
only_gold = df.where(df['Gold'] > 0)
only_gold.head()
```

```python
# Most statistical functions built into the Dataframe object 
# ignores values of NaN
only_gold['Gold'].count()
```

```python
df['Gold'].count()
```

```python
# dropna() drops all the rows/columns with NaN values
only_gold = only_gold.dropna()
only_gold.head()
```

```python
# Simpler way to do the same task
only_gold = df[df['Gold']>0]
only_gold.head()
```

```python
# Boolean Mask
# How many countries have won a gold medal in both summer and winter Olympics?
len(df[(df['Gold'] > 0) & (df['Gold.1'] > 0)])
```

```python
# Are there any countries that only won gold medals in winter Olympics but not summer ones?
df[(df['Gold.1'] > 0) & (df['Gold'] == 0)].head()
```

###  Indexing Dataframes

```python
# We already set axis0 to the Dataframe
# set_index() is a destructive method (it deletes current index)
df['country'] = df.index # create a new column called 'country' and set its value to current index
df = df.set_index('Gold') 
df.head()
```

```python
df = df.reset_index() # reset to default numerical index
df.head()
```
